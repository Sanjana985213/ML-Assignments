```python
for col in num_cols:
    print(col)
    print('Skew :', round(traffic_accidents[col].skew(), 2))
    plt.figure(figsize = (15, 4))
    plt.subplot(1, 2, 1)
    traffic_accidents[col].hist(grid=False)
    plt.ylabel('count')
    plt.subplot(1, 2, 2)
    sns.boxplot(x=traffic_accidents[col])
    plt.show()
```

```python
fig, axes = plt.subplots(3, 2, figsize = (18, 18))
fig.suptitle('Bar plot for all categorical variables in the dataset')
sns.countplot(ax = axes[0, 0], x = 'traffic_control_device', data = traffic_accidents, color = 'blue', 
              order = traffic_accidents['traffic_control_device'].value_counts().index);
sns.countplot(ax = axes[0, 1], x = 'weather_condition', data = traffic_accidents, color = 'blue', 
              order = traffic_accidents['weather_condition'].value_counts().index);
sns.countplot(ax = axes[1, 0], x = 'lighting_condition', data = traffic_accidents, color = 'blue', 
              order = traffic_accidents['lighting_condition'].value_counts().index);
sns.countplot(ax = axes[1, 1], x = 'first_crash_type', data = traffic_accidents, color = 'blue', 
              order = traffic_accidents['first_crash_type'].value_counts().index);
sns.countplot(ax = axes[2, 0], x = 'trafficway_type', data = traffic_accidents, color = 'blue', 
              order = traffic_accidents['trafficway_type'].head(20).value_counts().index);
sns.countplot(ax = axes[2, 1], x = 'roadway_surface_cond', data = traffic_accidents, color = 'blue', 
              order = traffic_accidents['roadway_surface_cond'].head(20).value_counts().index);
axes[1][1].tick_params(labelrotation=45);
axes[2][0].tick_params(labelrotation=90);
axes[2][1].tick_params(labelrotation=90);
```

```python
# Function for log transformation of the column
def log_transform(data,col):
    for colname in col:
        if (data[colname] == 1.0).all():
            data[colname + '_log'] = np.log(data[colname]+1)
        else:
            data[colname + '_log'] = np.log(data[colname])
    data.info()
```

```python
log_transform(traffic_accidents,['injuries_total','injuries_no_indication'])
```

```python
plt.figure(figsize=(13,17))
sns.pairplot(data=traffic_accidents.drop(['injuries_total','injuries_no_indication'],axis=1))
plt.show()
```

```python


fig, axarr = plt.subplots(4, 2, figsize=(12, 18))
traffic_accidents.groupby('most_severe_injury')['injuries_no_indication'].mean().sort_values(ascending=False).plot.bar(ax=axarr[0][0], fontsize=12)
axarr[0][0].set_title("most_severe_injury Vs injuries_no_indication", fontsize=18)
traffic_accidents.groupby('lighting_condition')['injuries_no_indication'].mean().sort_values(ascending=False).plot.bar(ax=axarr[0][1], fontsize=12)
axarr[0][1].set_title("lighting_condition Vs injuries_no_indication", fontsize=18)
traffic_accidents.groupby('first_crash_type')['injuries_no_indication'].mean().sort_values(ascending=False).plot.bar(ax=axarr[1][0], fontsize=12)
axarr[1][0].set_title("first_crash_type Vs injuries_no_indication", fontsize=18)
traffic_accidents.groupby('trafficway_type')['injuries_no_indication'].mean().sort_values(ascending=False).plot.bar(ax=axarr[1][1], fontsize=12)
axarr[1][1].set_title("trafficway_type Vs injuries_no_indication", fontsize=18)
traffic_accidents.groupby('alignment')['injuries_no_indication'].mean().sort_values(ascending=False).head(10).plot.bar(ax=axarr[2][0], fontsize=12)
axarr[2][0].set_title("alignment Vs injuries_no_indication", fontsize=18)
traffic_accidents.groupby('roadway_surface_cond')['injuries_no_indication'].mean().sort_values(ascending=False).head(10).plot.bar(ax=axarr[2][1], fontsize=12)
axarr[2][1].set_title("roadway_surface_cond Vs injuries_no_indication", fontsize=18)
traffic_accidents.groupby('road_defect')['injuries_no_indication'].mean().sort_values(ascending=False).plot.bar(ax=axarr[3][0], fontsize=12)
axarr[3][0].set_title("road_defect Vs injuries_no_indication", fontsize=18)
traffic_accidents.groupby('crash_type')['injuries_no_indication'].mean().sort_values(ascending=False).plot.bar(ax=axarr[3][1], fontsize=12)
axarr[3][1].set_title("crash_type Vs injuries_no_indication", fontsize=18)
plt.subplots_adjust(hspace=1.0)
plt.subplots_adjust(wspace=.5)
sns.despine()
```
